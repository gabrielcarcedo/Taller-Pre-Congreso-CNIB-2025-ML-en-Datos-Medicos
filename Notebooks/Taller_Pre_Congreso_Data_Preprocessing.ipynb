# Dataset ðŸ“‘

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np

# Cargar dataset
data = load_breast_cancer()
cancer_df = pd.DataFrame(data.data, columns=data.feature_names)

cancer_df
```

# **Data Preprocessing**

## Data Cleaning

image.png

```python
print("Missing values:\n", cancer_df.isnull().sum())
```

```python
# Simulate missing values
for i in range(300):
  cancer_df.iloc[np.random.randint(0, len(cancer_df)), np.random.randint(0, len(cancer_df.columns))] = np.nan

print(cancer_df.isnull().sum())
```

```python
# Imputation by mean

num_cols = cancer_df.select_dtypes(include=[np.number]).columns

cancer_df_clean = cancer_df.copy()
cancer_df_clean[num_cols] = cancer_df_clean[num_cols].fillna(cancer_df_clean[num_cols].mean())

print("After cleaning:\n", cancer_df_clean.isnull().sum())
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10,4))
sns.histplot(cancer_df['perimeter error'], kde=True, ax=ax[0], color='red')
ax[0].set_title('radius error with missing values')
sns.histplot(cancer_df_clean['perimeter error'], kde=True, ax=ax[1], color='green')
ax[1].set_title('radius error after imputation')
plt.show()
```

## Scaling

image.png

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# StandardScaler (mean=0, variance=1)
scaler_std = StandardScaler()
cancer_df_scaled = cancer_df_clean.copy()
cancer_df_scaled[num_cols] = scaler_std.fit_transform(cancer_df_clean[num_cols])

# MinMaxScaler (range: 0-1)
scaler_mm = MinMaxScaler()
cancer_df_mm = cancer_df_clean.copy()
cancer_df_mm[num_cols] = scaler_mm.fit_transform(cancer_df_clean[num_cols])

# Scatter plot para visualizar la transformaciÃ³n
plt.figure(figsize=(6,6))
plt.scatter(cancer_df_clean['mean area'], cancer_df_clean['worst radius'], alpha=0.5, label='Original')
plt.scatter(cancer_df_scaled['mean area'], cancer_df_scaled['worst radius'], alpha=0.5, label='Standarized')
plt.scatter(cancer_df_mm['mean area'], cancer_df_mm['worst radius'], alpha=0.5, label='MinMax')
plt.legend()
plt.xlabel('mean area')
plt.ylabel('worst radius')
plt.title('Scaling')
plt.grid(ls='--', alpha=0.5)
plt.show()
```

## Feature Selection

image.png

```python
from sklearn.feature_selection import SelectKBest, f_classif

selector = SelectKBest(f_classif, k=5)
selector.fit(cancer_df_scaled, data.target)

# Graficar scores
scores = selector.scores_
plt.figure(figsize=(10,4))
sns.barplot(x=data.feature_names, y=scores, order=data.feature_names[np.argsort(scores)])
plt.xticks(rotation=90)
plt.title("Features relevance (ANOVA F-score)")
plt.show()
```

```python
# Select top 5
cancer_df_top5 = cancer_df_scaled.iloc[:, selector.get_support()]

print("Forma original:", cancer_df_scaled.shape)
print("DespuÃ©s de selecciÃ³n:", cancer_df_top5.shape)
```

## Feature Extraction

image.png

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

cancer_df_PCA = pca.fit_transform(pd.get_dummies(cancer_df_top5))

plt.figure(figsize=(6,6))
plt.scatter(cancer_df_PCA[:,0], cancer_df_PCA[:,1], c=data.target, cmap='coolwarm', alpha=0.6)
plt.title('PCA with 2 components')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.colorbar(label='Cancer (0 = No, 1 = Yes)')
plt.grid(ls='--', alpha=0.5)
plt.show()
```

## Dimensionality reduction

image.png

```python
from sklearn.manifold import TSNE

# 2 dimensions with t-SNE
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
cancer_df_tsne = tsne.fit_transform(pd.get_dummies(cancer_df_top5))

plt.figure(figsize=(6,6))
plt.scatter(cancer_df_tsne[:,0], cancer_df_tsne[:,1], c=data.target, cmap='coolwarm', alpha=0.6)
plt.title('t-SNE con 2 dimensiones')
plt.xlabel('Dim 1')
plt.ylabel('Dim 2')
plt.colorbar(label='Cancer (0 = No, 1 = Yes)')
plt.grid(ls='--', alpha=0.5)
plt.show()
```

## Dataset split

image.png

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(cancer_df_clean, data.target,
                                                    test_size=0.2,
                                                    stratify=data.target,
                                                    random_state=1)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train,
                                                  test_size=0.2,
                                                  stratify=y_train,
                                                  random_state=1)

labels = ['Train', 'Validation', 'Test']
sizes = [len(y_train), len(y_val), len(y_test)]
plt.figure(figsize=(5,5))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#4c72b0','#dd8452', '#59a14f'])
plt.title('Dataset split â€“ Cancer')
plt.show()
```
